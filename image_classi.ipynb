```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
dig= load_digits()
```

```python
dig.data.shape
X = dig.data
```

```python
dig.target.shape
y = dig.target
```

```python
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(dig.data[0:10], dig.target[0:10])):
    plt.subplot(1, 10, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n'% label, fontsize = 20)
```

```python
plt.figure(figsize=(20,4))
for index, (image, label) in enumerate(zip(dig.data[0:5], dig.target[0:10])):
    plt.subplot(1, 5, index + 1)
    plt.imshow(np.reshape(image, (8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n'% label, fontsize = 20)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y,test_size=0.2, random_state=0)
```

```python
print("Train input data:", X_train.shape)
print("Test input data:", X_test.shape)
print("Train input data:", X_train.shape)
print("Test output data:", X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression().fit(X_train,y_train)
# log_reg.predict(X_test[0:50])
predictions=log_reg.predict(X_test)
predictions
```

```python
# accuracy test
score = log_reg.score(X_test, y_test)
print("the accuracy is:", score)
```

```python
# confusion matrix
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predictions)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f",linewidths=.5, square=True, cmap = 'Spectral');
plt.ylabel('actual output')
plt.xlabel("predicted output")
all_sample_title= 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15)
```
